# Probability Distributions

Probability distributions describe how likely different outcomes are in a random process.  
Before diving into machine learning, it's helpful to get an intuition for both **continuous distributions** (smooth curves over real numbers) and **discrete distributions** (probabilities assigned to specific values).



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import (
    norm, expon, beta, gamma, uniform, lognorm, chi2, laplace, weibull_min, t,
    bernoulli, binom, poisson, geom
)
```

### Visualizing Probability Distributions

The code below:
- Plots three continuous distributions:
  - **Normal (Gaussian)**: bell-shaped, centered at 0.
  - **Exponential**: models waiting times until the next event.
  - **Beta**: lives between 0 and 1, often used for modeling probabilities themselves.

```python
x_cont = np.linspace(-3, 3, 400)
x_unit = np.linspace(0, 1, 400)

plt.figure(figsize=(12, 8))

plt.plot(x_cont, norm.pdf(x_cont, 0, 1), label='Normal(0,1)')
plt.plot(x_cont[x_cont>=0], expon.pdf(x_cont[x_cont>=0], scale=1), label='Exponential(λ=1)')
# Beta (example α=2, β=5)
plt.plot(x_unit, beta.pdf(x_unit, 2, 5), label='Beta(2,5)')

plt.title('Continuous Distributions')
plt.legend()
plt.show()
```

## Probability Distribution Zoo 🦓📊

This code block visualizes a wide range of commonly used distributions:

**Continuous distributions**   
- **Gamma**: generalizes exponential, used for sums of waiting times.  
- **Uniform**: everything is equally likely in a range.  
- **Log-normal**: if the log of a variable is normal, the variable itself is log-normal.  
- **Chi-square**: appears in hypothesis testing and variance estimation.  
- **Laplace**: like a sharper-peaked normal, with heavier tails.  
- **Weibull**: popular in reliability analysis and survival modeling.  
- **Student’s t**: similar to normal, but with heavier tails (robust to outliers).  

**Discrete distributions**  
- **Bernoulli**: one yes/no trial.  
- **Binomial**: number of successes across fixed independent trials.  
- **Poisson**: counts of events in a fixed interval, with constant rate.  
- **Geometric**: number of trials until the first success.  
- **Categorical (Multinomial with n=1)**: probability over multiple categories.  

By plotting these, we can see how distributions differ in shape and support, and build the intuition needed for ML models, Bayesian reasoning, and statistical inference.


```python
# Continuous Distributions
x_cont = np.linspace(-5, 5, 500)
x_pos = np.linspace(0, 10, 500)
x_unit = np.linspace(0, 1, 500)

plt.figure(figsize=(12, 8))

plt.plot(x_pos, gamma.pdf(x_pos, a=2), label="Gamma(k=2, θ=1)", color='blue')
plt.plot(x_unit, uniform.pdf(x_unit, 0, 1), label="Uniform(0,1)", color='green')
plt.plot(x_pos, lognorm.pdf(x_pos, s=0.5), label="LogNormal(s=0.5)", color='orange')
plt.plot(x_pos, chi2.pdf(x_pos, df=3), label="Chi-square(df=3)", color='purple')
plt.plot(x_cont, laplace.pdf(x_cont, 0, 1), label="Laplace(0,1)", color='brown')
plt.plot(x_pos, weibull_min.pdf(x_pos, c=1.5), label="Weibull(c=1.5)", color='pink')
plt.plot(x_cont, t.pdf(x_cont, df=10), label="Student's t(df=10)", color='red')

plt.title("Various Continuous Distributions")
plt.legend()
plt.show()

# Discrete Distributions
plt.figure(figsize=(16, 12))

plt.subplot(3, 2, 1)
plt.stem([0, 1], bernoulli.pmf([0, 1], 0.5))
plt.title("Bernoulli(p=0.5)")

plt.subplot(3, 2, 2)
x_disc = np.arange(0, 11)
plt.stem(x_disc, binom.pmf(x_disc, n=10, p=0.3))
plt.title("Binomial(n=10,p=0.3)")

plt.subplot(3, 2, 3)
x_disc = np.arange(0, 15)
plt.stem(x_disc, poisson.pmf(x_disc, mu=4))
plt.title("Poisson(λ=4)")

plt.subplot(3, 2, 4)
x_disc = np.arange(1, 11)
plt.stem(x_disc, geom.pmf(x_disc, p=0.3))
plt.title("Geometric(p=0.3)")

plt.subplot(3, 2, 5)
pvals = [0.1, 0.3, 0.6]
x_disc = np.arange(len(pvals))
plt.bar(x_disc, pvals)
plt.xticks(x_disc, [f"cat{i}" for i in range(1, len(pvals)+1)])
plt.title("Categorical Distribution (Multinomial with n=1)")

plt.tight_layout()
plt.show()

```
